```python
import pandas as pd
from pathlib import Path
ROI_dir_name = 'ROI'

logs_output = pd.DataFrame()

ROI_path = Path(ROI_dir_name)
for file in ROI_path.glob('Pos*-*.txt'):
    data = pd.read_csv(file, delimiter='\s+')
    #delete the duplicate frame here?
    data.insert(0,'name',file.stem)
    logs_output = logs_output.append(data, ignore_index=True)
        
logs_output['frame'] = (logs_output['Label'].str.extract(r'tif.[0-9]{1,4}-[0-9]{1,4}-[0-9]{1,4}:(?P<frame>\d{1,3})')).astype(int)
logs_output[['Position','cell_id']] = logs_output['name'].str.extract(r'(?P<Pos>Pos[0-9]{1,3})-(?P<id>[0-9]{1,3})')
logs_output_data = logs_output.drop_duplicates(subset=['name', 'frame'],keep='last',ignore_index=True)

duplicateRowsDF = logs_output[logs_output.duplicated(subset=['name', 'frame'],keep='last')]
#duplicateRowsDF = logs_output[logs_output.duplicated(subset=['name', 'frame'],keep=False)]
print("Duplicate Rows based on a single column are:", duplicateRowsDF, sep='\n')
```
